```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import os
import statsmodels.api as sm
from scipy import signal
from scipy import stats
import pandas as pd
```

### Time series analysis 

```python
# Read data
enso_ts = pd.read_csv('./enso_ond.csv')
jet_ts = pd.read_csv('./jet_ond.csv')
spv_ts = pd.read_csv('./spv_ond.csv')

# Set 'year' as index
enso_ts.set_index('year', inplace=True)
jet_ts.set_index('year', inplace=True)
spv_ts.set_index('year', inplace=True)

# Create a single figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Plot ENSO
axes[0].plot(enso_ts.index, enso_ts['enso'], label='ENSO')
axes[0].set_title('ENSO Time Series')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('ENSO Index')
axes[0].legend()

# Plot Jet
axes[1].plot(jet_ts.index, jet_ts['jet'], label='Jet', color='orange')
axes[1].set_title('Jet Time Series')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Jet Index')
axes[1].legend()

# Plot SPV
axes[2].plot(spv_ts.index, spv_ts['spv'], label='SPV', color='green')
axes[2].set_title('SPV Time Series')
axes[2].set_xlabel('Year')
axes[2].set_ylabel('SPV Index')
axes[2].legend()

# Optimise layout
plt.tight_layout()

# Show the figure
plt.show()
```
